```python
from collections import deque

class Graph:
    def __init__(self, adjac_lis): # Sửa lỗi gõ: __init__
        self.adjac_lis = adjac_lis

    def get_neighbors(self, v): # Sửa lỗi gõ: get_neighbors
        return self.adjac_lis[v]

    # Hàm heuristic: Khoảng cách ước lượng từ n đến đích
    def h(self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return H[n]

    def a_star_algorithm(self, start, stop):
        # open_lst: Nút đã phát hiện nhưng chưa duyệt xong
        # closed_lst: Nút đã duyệt xong
        open_lst = set([start])
        closed_lst = set([])

        # g (poo) chứa chi phí thực tế từ start đến nút hiện tại
        g = {}
        g[start] = 0

        # par chứa ánh xạ nút cha để truy hồi đường đi
        par = {}
        par[start] = start

        while len(open_lst) > 0:
            n = None

            # Tìm nút trong open_lst có f(n) = g(n) + h(n) nhỏ nhất
            for v in open_lst:
                if n == None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n == None:
                print('Path does not exist!')
                return None

            # Nếu tìm thấy đích
            if n == stop:
                reconst_path = []
                while par[n] != n:
                    reconst_path.append(n)
                    n = par[n]
                reconst_path.append(start)
                reconst_path.reverse()
                print('Path found: {}'.format(reconst_path))
                return reconst_path

            # Duyệt các nút lân cận
            for (m, weight) in self.get_neighbors(n):
                # Nếu m chưa có trong cả 2 danh sách
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    g[m] = g[n] + weight
                else:
                    # Nếu tìm thấy đường đi đến m ngắn hơn đường cũ
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        par[m] = n

                        # Nếu m đã nằm trong closed_lst, chuyển lại sang open_lst để cập nhật các con của nó
                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)

            # Xóa n khỏi open_lst và thêm vào closed_lst sau khi duyệt xong neighbors
            open_lst.remove(n)
            closed_lst.add(n)

        print('Path does not exist!')
        return None

# Khởi tạo dữ liệu
adjac_lis = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)],
    'D': [] # Cần thêm nút D trống để không bị lỗi KeyError
}

graph1 = Graph(adjac_lis)
graph1.a_star_algorithm('A', 'D')
```

```python
#Cài đặt thuật giải A* tìm đường đi giữa 2 đỉnh bất kỳ trong đồ thị.
class Graph:
    def __init__(self, adjacency_list):
        """Khởi tạo đồ thị với danh sách kề"""
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list.get(v, [])

    def h(self, n, goal):
        """
        Hàm Heuristic (h): Ước lượng khoảng cách từ n đến đích.
        Trong thực tế, đây có thể là khoảng cách đường chim bay (Euclidean).
        Ở đây dùng một bộ giá trị mẫu.
        """
        H = {
            'A': 10, 'B': 8, 'C': 5, 'D': 7, 'E': 3, 'F': 0
        }
        return H.get(n, 0)

    def a_star_algorithm(self, start_node, stop_node):
        # open_list: Các nút đã tìm thấy nhưng chưa duyệt hết (Dùng set để tối ưu tốc độ)
        open_list = set([start_node])
        # closed_list: Các nút đã duyệt xong
        closed_list = set([])

        # g: Chi phí thực tế từ điểm bắt đầu đến nút hiện tại
        g = {start_node: 0}

        # parents: Lưu vết đường đi (mapping con -> cha)
        parents = {start_node: start_node}

        while len(open_list) > 0:
            n = None

            # Tìm nút n trong open_list có f(n) = g(n) + h(n) thấp nhất
            for v in open_list:
                if n == None or g[v] + self.h(v, stop_node) < g[n] + self.h(n, stop_node):
                    n = v

            if n == None:
                print('Không tìm thấy đường đi!')
                return None

            # Nếu n là đích, tiến hành truy hồi đường đi
            if n == stop_node:
                reconst_path = []
                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]
                reconst_path.append(start_node)
                reconst_path.reverse()

                print(f'Đường đi từ {start_node} đến {stop_node}: {reconst_path}')
                return reconst_path

            # Duyệt tất cả các nút lân cận của n
            for (m, weight) in self.get_neighbors(n):
                # Nếu m chưa có trong open_list và closed_list
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    # Nếu đã tồn tại, kiểm tra xem đường đi mới qua n có ngắn hơn không
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n

                        # Nếu m nằm trong closed_list, chuyển nó lại open_list để cập nhật con của nó
                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            # Sau khi duyệt hết lân cận, chuyển n sang danh sách đóng
            open_list.remove(n)
            closed_list.add(n)

        print('Không tìm thấy đường đi!')
        return None

# --- Chạy thử nghiệm ---
# Khai báo đồ thị: Nút: [(Nút kề, Trọng số cạnh), ...]
data = {
    'A': [('B', 1), ('C', 4)],
    'B': [('C', 2), ('D', 5), ('E', 12)],
    'C': [('E', 3)],
    'D': [('F', 2)],
    'E': [('F', 3)],
    'F': []
}

graph = Graph(data)
graph.a_star_algorithm('A', 'F')
```
