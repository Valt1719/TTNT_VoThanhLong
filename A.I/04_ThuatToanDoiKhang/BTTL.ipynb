```python
#Cài đặt thuật toán cắt tỉa Alpha-beta với ứng dụng
#vào bài toán TicTacToe với ma trận 3x3.
import os, math

def GetWinner(board):
    """
    Returns the winner in the current board if there is one, otherwise it returns None.
    """
    # horizontal
    if board[0] == board[1] and board[1] == board[2]:
        return board[0]
    elif board[3] == board[4] and board[4] == board[5]:
        return board[3]
    elif board[6] == board[7] and board[7] == board[8]:
        return board[6]
    # vertical
    elif board[0] == board[3] and board[3] == board[6]:
        return board[0]
    elif board[1] == board[4] and board[4] == board[7]:
        return board[1]
    elif board[2] == board[5] and board[5] == board[8]:
        return board[2]
    # diagonal
    elif board[0] == board[4] and board[4] == board[8]:
        return board[0]
    elif board[2] == board[4] and board[4] == board[6]:
        return board[2]

def PrintBoard(board):
    """
    Clears the console and prints the current board.
    """
    os.system('cls' if os.name=='nt' else 'clear')
    print(f'''
    {board[0]}|{board[1]}|{board[2]}
    {board[3]}|{board[4]}|{board[5]}
    {board[6]}|{board[7]}|{board[8]}
    ''')

def GetAvailableCells(board):
    """
    Returns a list of indices containing all available cells in a board.
    """
    available = list()
    for cell in board:
        if cell != "X" and cell != "O":
            available.append(cell)
    return available

def minimax(position, depth, alpha, beta, isMaximizing):
    """
    The AI algorithm responsible for choosing the best move. Returns best value of a move.
    """
    # evaluate current board: if maximizing player won -> return 10
    #                         if minimizing player won -> return -10
    #                         if no one is winning (tie) -> return 0

    # NOTE: Even though the following AI plays perfectly, it might
    #       choose to make a move which will result in a slower victory
    #       or a faster loss. Lets take an example and explain it
    #       Assume that there are 2 possible ways for X to win the game from a give board state.
    #       Move A : X can win in 2 move
    #       Move B : X can win in 4 moves
    #       Our evaluation will return a value of +10 for both moves A and B. Even though the move A
    #       is better because it ensures a faster victory, our AI may choose B sometimes. To overcome
    #       this problem we subtract the depth value from the evaluated score. This means that in case
    #       of a victory it will choose a the victory which takes least number of moves and in case of
    #       a loss it will try to prolong the game and play as many moves as possible.
    winner = GetWinner(position)
    if winner != None:
        return 10 - depth if winner == "X" else -10 + depth
    if len(GetAvailableCells(position)) == 0:
        return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            Eval = minimax(position, depth + 1, alpha, beta, False)
            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return maxEval
    else:
        minEval = +math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"
            Eval = minimax(position, depth + 1, alpha, beta, True)
            minEval = min(minEval, Eval)
            beta = min(beta, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return minEval

def FindBestMove(currentPosition, AI):
    """
    This will return the best possible move for the player.
    Will Traverse all cells, evaluate minimax function for all empty cells.
    And return the cell with optimal value.
    Parameters:
        currentPosition (list): The current board to find best move for.
        AI (str): The AI Player ("X" or "O").
    Returns:
        int: Index of best move for the current position.
    """
    bestVal = -math.inf if AI == "X" else +math.inf
    bestMove = -1
    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        moveVal = minimax(currentPosition, 0, -math.inf, +math.inf, False if AI == "X" else True)
        currentPosition[cell - 1] = cell
        if (AI == "X" and moveVal > bestVal):
            bestMove = cell
            bestVal = moveVal
        elif (AI == "O" and moveVal < bestVal):
            bestMove = cell
            bestVal = moveVal
    return bestMove

def main():
    player = input("Play as X or O? ").strip().upper()
    AI = "O" if player == "X" else "X"
    currentGame = [*range(1, 10)]
    # X always starts first.
    currentTurn = "X"
    counter = 0
    while True:
        if currentTurn == AI:
            # NOTE: if the AI starts first, it'll always choose index 0 so to save time you could play it.
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player
        elif currentTurn == player:
            PrintBoard(currentGame)
            while True:
                humanInput = int(input("Enter Number: ").strip())
                if humanInput in currentGame:
                    currentGame[humanInput - 1] = player
                    currentTurn = AI
                    break
                else:
                    PrintBoard(currentGame)
                    print("Cell Not Available.")
        if GetWinner(currentGame) != None:
            PrintBoard(currentGame)
            print(f"{GetWinner(currentGame)} WON!!!")
            break
        counter += 1
        if GetWinner(currentGame) == None and counter == 9:
            PrintBoard(currentGame)
            print("Tie.")
            break

if __name__ == "__main__":
    main()
```

```python
#caro NxN
import math
import os

class TicTacToe:
    def __init__(self, size=5, win_con=4):
        self.size = size
        self.win_con = win_con
        self.board = [[None for _ in range(size)] for _ in range(size)]
        self.ai_player = 'X'
        self.human_player = 'O'

    def print_board(self):
        os.system('cls' if os.name == 'nt' else 'clear')
        # In tiêu đề cột
        print("   " + " ".join([f"{i:2}" for i in range(self.size)]))
        for r in range(self.size):
            row_str = f"{r:2} "
            for c in range(self.size):
                symbol = self.board[r][c] if self.board[r][c] else "."
                row_str += f" {symbol} "
            print(row_str)

    def check_winner(self, b):
        # Kiểm tra nhanh xem ai thắng trên bàn cờ b
        for r in range(self.size):
            for c in range(self.size):
                if b[r][c] is None: continue
                # Kiểm tra 4 hướng: ngang, dọc, chéo xuôi, chéo ngược
                for dr, dc in [(0, 1), (1, 0), (1, 1), (1, -1)]:
                    count = 0
                    for i in range(self.win_con):
                        nr, nc = r + dr*i, c + dc*i
                        if 0 <= nr < self.size and 0 <= nc < self.size and b[nr][nc] == b[r][c]:
                            count += 1
                        else: break
                    if count == self.win_con: return b[r][c]
        return None

    def is_full(self):
        return all(cell is not None for row in self.board for cell in row)

    def evaluate(self):
        """
        Hàm lượng giá đơn giản:
        Càng nhiều quân xếp hàng thì điểm càng cao.
        """
        winner = self.check_winner(self.board)
        if winner == self.ai_player: return 1000
        if winner == self.human_player: return -1000
        return 0

    def get_moves(self):
        # Chỉ lấy các ô trống gần các ô đã có quân để tăng tốc
        moves = []
        for r in range(self.size):
            for c in range(self.size):
                if self.board[r][c] is None:
                    # Tối ưu: Chỉ chọn ô có lân cận trong bán kính 1 ô
                    has_neighbor = False
                    for dr in range(-1, 2):
                        for dc in range(-1, 2):
                            nr, nc = r + dr, c + dc
                            if 0 <= nr < self.size and 0 <= nc < self.size and self.board[nr][nc]:
                                has_neighbor = True; break
                        if has_neighbor: break
                    if has_neighbor or (r == self.size//2 and c == self.size//2):
                        moves.append((r, c))
        return moves

    def alpha_beta(self, depth, alpha, beta, is_max):
        score = self.evaluate()
        if score != 0 or depth == 0 or self.is_full():
            return score

        moves = self.get_moves()
        if is_max:
            max_val = -math.inf
            for r, c in moves:
                self.board[r][c] = self.ai_player
                val = self.alpha_beta(depth - 1, alpha, beta, False)
                self.board[r][c] = None
                max_val = max(max_val, val)
                alpha = max(alpha, max_val)
                if beta <= alpha: break
            return max_val
        else:
            min_val = math.inf
            for r, c in moves:
                self.board[r][c] = self.human_player
                val = self.alpha_beta(depth - 1, alpha, beta, True)
                self.board[r][c] = None
                min_val = min(min_val, val)
                beta = min(beta, min_val)
                if beta <= alpha: break
            return min_val

    def find_best_move(self):
        best_val = -math.inf
        best_move = None
        # Depth = 3 để đảm bảo tốc độ cho bàn cờ lớn
        for r, c in self.get_moves():
            self.board[r][c] = self.ai_player
            move_val = self.alpha_beta(3, -math.inf, math.inf, False)
            self.board[r][c] = None
            if move_val > best_val:
                best_val = move_val
                best_move = (r, c)
        return best_move

def main():
    print("--- TIC TAC TOE N x N (ALPHA-BETA) ---")
    n = int(input("Nhập kích thước bàn cờ (ví dụ 5): "))
    win = int(input("Nhập số quân liên tiếp để thắng (ví dụ 4): "))

    game = TicTacToe(size=n, win_con=win)
    current_turn = 'O' # Người đi trước

    while True:
        game.print_board()

        if current_turn == game.human_player:
            print("\nLượt của bạn (O):")
            try:
                r, c = map(int, input("Nhập hàng và cột (ví dụ: 1 2): ").split())
                if game.board[r][c] is not None:
                    print("Ô đã có quân! Nhập lại."); continue
                game.board[r][c] = game.human_player
                current_turn = game.ai_player
            except:
                print("Lỗi nhập liệu!"); continue
        else:
            print("\nAI (X) đang tính toán...")
            move = game.find_best_move()
            if move:
                game.board[move[0]][move[1]] = game.ai_player
            current_turn = game.human_player

        winner = game.check_winner(game.board)
        if winner or game.is_full():
            game.print_board()
            if winner: print(f"\nKẾT THÚC: {winner} THẮNG!")
            else: print("\nHÒA!")
            break

if __name__ == "__main__":
    main()
```

```python
import math
import os

# --- HẰNG SỐ & ĐÁNH GIÁ QUÂN CỜ ---
# Chữ hoa: Đỏ (Maximizer) | Chữ thường: Đen (Minimizer)
PIECE_VALUES = {
    'K': 10000, 'R': 900, 'C': 450, 'N': 400, 'B': 200, 'A': 200, 'P': 100,
    'k': -10000, 'r': -900, 'c': -450, 'n': -400, 'b': -200, 'a': -200, 'p': -100
}

INITIAL_BOARD = [
    ['r', 'n', 'b', 'a', 'k', 'a', 'b', 'n', 'r'],
    ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
    ['.', 'c', '.', '.', '.', '.', '.', 'c', '.'],
    ['p', '.', 'p', '.', 'p', '.', 'p', '.', 'p'],
    ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
    ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
    ['P', '.', 'P', '.', 'P', '.', 'P', '.', 'P'],
    ['.', 'C', '.', '.', '.', '.', '.', 'C', '.'],
    ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
    ['R', 'N', 'B', 'A', 'K', 'A', 'B', 'N', 'R']
]

class Board:
    def __init__(self):
        self.state = [row[:] for row in INITIAL_BOARD]

    def print_board(self):
        os.system('cls' if os.name == 'nt' else 'clear')
        print("    0 1 2 3 4 5 6 7 8 (Cột)")
        print("  ---------------------")
        for r in range(10):
            row_str = f"{r} | "
            for c in range(9):
                row_str += self.state[r][c] + " "
            print(row_str)
        print("  ---------------------")

    def evaluate(self):
        """Hàm lượng giá: Tổng điểm Đỏ + Tổng điểm Đen"""
        score = 0
        for r in range(10):
            for c in range(9):
                piece = self.state[r][c]
                if piece != '.':
                    score += PIECE_VALUES[piece]
                    # Điểm cộng thêm cho Tốt Đỏ qua sông
                    if piece == 'P' and r < 5: score += 50
                    # Điểm cộng thêm cho Tốt Đen qua sông
                    if piece == 'p' and r > 4: score -= 50
        return score

    def is_red(self, piece): return piece.isupper()
    def is_black(self, piece): return piece.islower() and piece != '.'

    def get_moves(self, is_red_turn):
        """Sinh ra tất cả nước đi hợp lệ cho phe hiện tại"""
        moves = []
        for r in range(10):
            for c in range(9):
                piece = self.state[r][c]
                if piece == '.': continue
                if (is_red_turn and self.is_red(piece)) or (not is_red_turn and self.is_black(piece)):
                    moves.extend(self.get_piece_moves(r, c, piece))
        return moves

    def get_piece_moves(self, r, c, piece):
        moves = []
        p_type = piece.upper()
        enemy_func = self.is_black if self.is_red(piece) else self.is_red

        def add_if_valid(nr, nc):
            if 0 <= nr < 10 and 0 <= nc < 9:
                target = self.state[nr][nc]
                if target == '.' or enemy_func(target):
                    moves.append(((r, c), (nr, nc)))
                    return target != '.' # Trả về True nếu ăn quân
            return False

        # 1. XE (Chariot)
        if p_type == 'R':
            for dr, dc in [(-1,0), (1,0), (0,-1), (0,1)]:
                nr, nc = r + dr, c + dc
                while 0 <= nr < 10 and 0 <= nc < 9:
                    if self.state[nr][nc] == '.':
                        moves.append(((r, c), (nr, nc)))
                    elif enemy_func(self.state[nr][nc]):
                        moves.append(((r, c), (nr, nc)))
                        break
                    else: break
                    nr, nc = nr + dr, nc + dc

        # 2. PHÁO (Cannon)
        elif p_type == 'C':
            for dr, dc in [(-1,0), (1,0), (0,-1), (0,1)]:
                nr, nc = r + dr, c + dc
                jumped = False
                while 0 <= nr < 10 and 0 <= nc < 9:
                    target = self.state[nr][nc]
                    if not jumped:
                        if target == '.': moves.append(((r, c), (nr, nc)))
                        else: jumped = True # Gặp ngòi
                    else:
                        if target != '.':
                            if enemy_func(target): moves.append(((r, c), (nr, nc)))
                            break
                    nr, nc = nr + dr, nc + dc

        # 3. MÃ (Horse) - Xét cản chân
        elif p_type == 'N':
            horse_moves = [
                (-2, -1, -1, 0), (-2, 1, -1, 0), (2, -1, 1, 0), (2, 1, 1, 0),
                (-1, -2, 0, -1), (1, -2, 0, -1), (-1, 2, 0, 1), (1, 2, 0, 1)
            ]
            for dr, dc, br, bc in horse_moves:
                if 0 <= r + br < 10 and 0 <= c + bc < 9 and self.state[r + br][c + bc] == '.':
                    add_if_valid(r + dr, c + dc)

        # 4. TỐT (Pawn)
        elif p_type == 'P':
            dir_r = -1 if self.is_red(piece) else 1
            add_if_valid(r + dir_r, c)
            # Nếu đã qua sông (Đỏ r < 5, Đen r > 4), được đi ngang
            if (self.is_red(piece) and r < 5) or (self.is_black(piece) and r > 4):
                add_if_valid(r, c - 1)
                add_if_valid(r, c + 1)

        # 5. TƯỚNG (King) & 6. SĨ (Advisor) & 7. TƯỢNG (Elephant)
        # Giới hạn Cửu cung và Sông. (Viết rút gọn để tối ưu code)
        elif p_type in ['K', 'A', 'B']:
            if p_type == 'K': dirs = [(-1,0), (1,0), (0,-1), (0,1)]
            elif p_type == 'A': dirs = [(-1,-1), (-1,1), (1,-1), (1,1)]
            else: dirs = [(-2,-2), (-2,2), (2,-2), (2,2)] # Tượng

            for dr, dc in dirs:
                nr, nc = r + dr, c + dc

                # Ràng buộc Tướng/Sĩ trong cung
                if p_type in ['K', 'A']:
                    if not (3 <= nc <= 5): continue
                    if self.is_red(piece) and not (7 <= nr <= 9): continue
                    if self.is_black(piece) and not (0 <= nr <= 2): continue
                # Ràng buộc Tượng không qua sông và không bị cản
                if p_type == 'B':
                    if self.is_red(piece) and nr < 5: continue
                    if self.is_black(piece) and nr > 4: continue

                    # New check for intermediate square validity
                    mid_r, mid_c = r + dr//2, c + dc//2
                    if not (0 <= mid_r < 10 and 0 <= mid_c < 9): # Check if intermediate cell is within bounds
                        continue

                    if self.state[mid_r][mid_c] != '.': continue # Cản mắt Tượng

                add_if_valid(nr, nc)

        return moves

class XiangqiAI:
    def __init__(self, depth=3):
        self.depth = depth

    def alpha_beta(self, board, depth, alpha, beta, is_maximizing):
        # Kiểm tra kết thúc sớm (Mất tướng)
        k_count = sum(row.count('K') for row in board.state)
        k_count_black = sum(row.count('k') for row in board.state)
        if k_count == 0: return -10000 + depth # Đỏ thua
        if k_count_black == 0: return 10000 - depth # Đen thua

        if depth == 0:
            return board.evaluate()

        moves = board.get_moves(is_maximizing)

        # Move Ordering (Ưu tiên các nước ăn quân để cắt tỉa Alpha-Beta nhanh hơn)
        # Sắp xếp các nước đi: Nước nào ô đích đang có quân đứng -> xếp lên đầu
        moves.sort(key=lambda m: 1 if board.state[m[1][0]][m[1][1]] != '.' else 0, reverse=True)

        if not moves: # Hết nước đi (bị bí)
            return -10000 if is_maximizing else 10000

        if is_maximizing:
            max_eval = -math.inf
            for start, end in moves:
                # Lưu trạng thái để undo
                target_piece = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = board.state[start[0]][start[1]]
                board.state[start[0]][start[1]] = '.'

                eval_val = self.alpha_beta(board, depth - 1, alpha, beta, False)

                # Undo
                board.state[start[0]][start[1]] = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = target_piece

                max_eval = max(max_eval, eval_val)
                alpha = max(alpha, eval_val)
                if beta <= alpha: break
            return max_eval
        else:
            min_eval = math.inf
            for start, end in moves:
                target_piece = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = board.state[start[0]][start[1]]
                board.state[start[0]][start[1]] = '.'

                eval_val = self.alpha_beta(board, depth - 1, alpha, beta, True)

                board.state[start[0]][start[1]] = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = target_piece

                min_eval = min(min_eval, eval_val)
                beta = min(beta, eval_val)
                if beta <= alpha: break
            return min_eval

    def find_best_move(self, board, is_red):
        best_move = None
        alpha = -math.inf
        beta = math.inf

        moves = board.get_moves(is_red)
        if not moves: return None

        if is_red:
            best_val = -math.inf
            for start, end in moves:
                target = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = board.state[start[0]][start[1]]
                board.state[start[0]][start[1]] = '.'

                move_val = self.alpha_beta(board, self.depth - 1, alpha, beta, False)

                board.state[start[0]][start[1]] = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = target

                if move_val > best_val:
                    best_val = move_val
                    best_move = (start, end)
        else:
            best_val = math.inf
            for start, end in moves:
                target = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = board.state[start[0]][start[1]]
                board.state[start[0]][start[1]] = '.'

                move_val = self.alpha_beta(board, self.depth - 1, alpha, beta, True)

                board.state[start[0]][start[1]] = board.state[end[0]][end[1]]
                board.state[end[0]][end[1]] = target

                if move_val < best_val:
                    best_val = move_val
                    best_move = (start, end)

        return best_move

# --- VÒNG LẶP GAME CHÍNH ---
def main():
    board = Board()
    ai = XiangqiAI(depth=3) # Độ sâu AI (3 là mượt, 4 sẽ hơi chậm trên Python thuần)

    print("CHÀO MỪNG ĐẾN VỚI CỜ TƯỚNG AI!")
    print("Bạn cầm quân Đỏ (chữ Hoa), AI cầm quân Đen (chữ thường).")
    input("Nhấn Enter để bắt đầu...")

    while True:
        board.print_board()

        # Kiểm tra thắng thua
        k_red = sum(row.count('K') for row in board.state)
        k_black = sum(row.count('k') for row in board.state)
        if k_red == 0: print("AI THẮNG!"); break
        if k_black == 0: print("BẠN THẮNG!"); break

        # Lượt người chơi (Đỏ)
        while True:
            try:
                print("\nLượt của bạn. Định dạng: Hàng_Nguồn Cột_Nguồn Hàng_Đích Cột_Đích (vd: 9 1 7 2 để nhảy Mã)")
                move_str = input("Nhập nước đi: ").split()
                if len(move_str) != 4: continue
                sr, sc, er, ec = map(int, move_str)

                # Kiểm tra hợp lệ cơ bản
                piece = board.state[sr][sc]
                if not board.is_red(piece):
                    print("Đó không phải quân của bạn!")
                    continue

                valid_moves = board.get_piece_moves(sr, sc, piece)
                if ((sr, sc), (er, ec)) in valid_moves:
                    board.state[er][ec] = board.state[sr][sc]
                    board.state[sr][sc] = '.'
                    break
                else:
                    print("Nước đi sai luật cờ tướng!")
            except Exception as e:
                print("Lỗi nhập liệu. Vui lòng nhập 4 số cách nhau bởi khoảng trắng.")

        board.print_board()

        k_black = sum(row.count('k') for row in board.state)
        if k_black == 0: print("BẠN THẮNG!"); break

        # Lượt AI (Đen)
        print("\nAI đang suy nghĩ...")
        best_move = ai.find_best_move(board, is_red=False)
        if best_move:
            start, end = best_move
            board.state[end[0]][end[1]] = board.state[start[0]][start[1]]
            board.state[start[0]][start[1]] = '.'
            print(f"AI đã đi từ {start} đến {end}")
        else:
            print("AI hết nước đi. BẠN THẮNG!")
            break

if __name__ == "__main__":
    main()
```
