```python
import copy
from heapq import heappush, heappop

n = 3
rows = [1, 0, -1, 0] # Down, Left, Up, Right
cols = [0, -1, 0, 1]

class PriorityQueue:
    def __init__(self):
        self.heap = []
    def push(self, key):
        heappush(self.heap, key)
    def pop(self):
        return heappop(self.heap)
    def empty(self):
        return not self.heap

class Node:
    def __init__(self, parent, mats, empty_tile_posi, costs, levels):
        self.parent = parent
        self.mats = mats
        self.empty_tile_posi = empty_tile_posi
        self.costs = costs
        self.levels = levels

    # Hàm so sánh để Priority Queue biết chọn node có costs thấp nhất
    def __lt__(self, nxt):
        return self.costs < nxt.costs

def calculateCost(mats, final):
    count = 0
    for i in range(n):
        for j in range(n):
            if mats[i][j] != 0 and mats[i][j] != final[i][j]:
                count += 1
    return count

def newNodes(mats, empty_tile_posi, new_empty_tile_posi, levels, parent, final):
    new_mats = copy.deepcopy(mats)
    x1, y1 = empty_tile_posi
    x2, y2 = new_empty_tile_posi
    # Hoán đổi vị trí ô trống (Swap tiles)
    new_mats[x1][y1], new_mats[x2][y2] = new_mats[x2][y2], new_mats[x1][y1]

    # Cost = số ô sai vị trí + số bước đã đi (g + h)
    costs = calculateCost(new_mats, final)
    return Node(parent, new_mats, new_empty_tile_posi, costs + levels, levels)

def printMatrix(mats):
    for row in mats:
        print(" ".join(map(str, row)))
    print()

def isSafe(x, y):
    return 0 <= x < n and 0 <= y < n

def printPath(root):
    if root is None: return
    printPath(root.parent)
    printMatrix(root.mats)

def solve(initial, empty_tile_posi, final):
    pq = PriorityQueue()
    cost = calculateCost(initial, final)
    root = Node(None, initial, empty_tile_posi, cost, 0)
    pq.push(root)

    while not pq.empty():
        minimum = pq.pop()

        # Nếu số ô sai vị trí = số bước đã đi (tức là h=0), ta đã tới đích
        if calculateCost(minimum.mats, final) == 0:
            print("--- Đường đi tìm được ---")
            printPath(minimum)
            return

        # Tạo tất cả các con khả thi (4 hướng)
        for i in range(4):
            new_tile_posi = [
                minimum.empty_tile_posi[0] + rows[i],
                minimum.empty_tile_posi[1] + cols[i],
            ]

            if isSafe(new_tile_posi[0], new_tile_posi[1]):
                child = newNodes(minimum.mats, minimum.empty_tile_posi,
                                 new_tile_posi, minimum.levels + 1, minimum, final)
                pq.push(child)

# Chạy chương trình
initial = [[1, 2, 3], [5, 6, 0], [7, 8, 4]]
final = [[1, 2, 3], [5, 8, 6], [0, 7, 4]]
empty_tile_posi = [1, 2]

solve(initial, empty_tile_posi, final)
```
