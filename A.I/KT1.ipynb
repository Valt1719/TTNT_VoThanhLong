# CÂU 1 (2 ĐIỂM) – Quay Lui: Sinh Hoán Vị & Derangement

Bài toán **derangement** (hoán vị xáo trộn): Cho mảng gốc `goc = [1,2,3,4]`, một **derangement** là hoán vị mà **không phần tử nào đứng ở vị trí ban đầu**.

> *Ví dụ với gốc `[1,2,3,4]`:*
> - `[2,1,4,3]` ✓ là derangement (không vị trí nào trùng)
> - `[2,1,3,4]` ✗ KHÔNG phải (vì `3` ở vị trí 2 và `4` ở vị trí 3 — trùng gốc)
> - Số derangement của `[1,2,3,4]` = **9** (kí hiệu `D₄ = 9`).

**b) (0.3đ)** Chạy `hoan_vi(arr=[1,2,3,4])` và đếm tổng số hoán vị in ra. Verify = `4! = 24`.

**c) (0.6đ)** Gọi `derangement([1,2,3,4], [1,2,3,4], 0, [])`. In tất cả derangement và đếm. Kiểm tra = **9**. In thêm số lần gọi đệ quy `dem_d`.

**d) (0.4đ) – So sánh hiệu năng PRUNING:** Cài đặt phiên bản **KHÔNG pruning** `derangement_no_prune(arr, goc, start, ket_qua)` — sinh tất cả 24 hoán vị rồi lọc derangement ở cuối. So sánh số lần gọi đệ quy với phiên bản có pruning. Giải thích bằng comment vì sao pruning hiệu quả hơn.

````python
# CAU 1a – Nhin anh phia tren, go lai toan bo code va dien 7 cho trong
# Khong duoc copy tu nguon khac
# ============================================================

dem_goi = 0

def hoan_vi(arr, start=0):
  """Sinh tat ca hoan vi cua arr (in-place swap)"""
  global dem_goi
  dem_goi += 1
  if start == len(arr): #Trong1
    print(arr.copy())
    return
  for i in range(start, len(arr)): #Trong2
    arr[start], arr[i] = arr[i], arr[start] #Trong3
    hoan_vi(arr, start+1) #Trong4
    #QUAY LUI: undo swap (tra lai trang thai cu)
    arr[start], arr[i] = arr[i], arr[start] #Trong5

# hoan vi khong co phan tu nao o vi tri goc
# vi du goc=[1,2,3,4]: [2,1,4,3] OK | [2,1,3,4] KO (3 va 4 con trung goc)
dem_d =0

def derangement(arr, goc, start, ket_qua):
  """Sinh tat ca derangement voi Pruning som(Cat nhanh som)"""
  global dem_d
  dem_d += 1
  if start == len(arr):
    ket_qua.append(arr.copy())
    return
  for i in range(start, len(arr)):
    #pruning: phan tu arr[i] se trung vi tri goc -> bo qua nhanh nay
    if arr[i] == goc[i]: #Trong6
      continue
    arr[start], arr[i] = arr[i], arr[start]
    derangement(arr, goc, start+1, ket_qua) #Trong7
    arr[start], arr[i] = arr[i], arr[start]
````

````python
# CAU 2d – Cai dat UCS (Uniform Cost Search) va so sanh voi A*
# UCS = A* voi h(n) = 0 cho moi node (tuong duong Dijkstra)
# ============================================================

def ucs(graph, start, goal):
    """
    Uniform Cost Search.
    Chi dung g(n) - khong dung heuristic.
    Tra ve (duong_di, tong_chi_phi, so_node_expand).
    """
    # TODO: Cai dat UCS bang priority queue, sap xep theo g(n)
    pass
    open_list = [(0, start, [start])]
    closed_list = set()
    n_exp = 0

    while open_list:
        g, current, path = heapq.heappop(open_list)

        if current in closed_list:
            continue

        n_exp += 1
        if current == goal:
            return path, g, n_exp

        closed_list.add(current)

        for neighbor, weight in graph.get(current, []):
            if neighbor not in closed_list:
                new_g = g + weight
                heapq.heappush(open_list, (new_g, neighbor, path + [neighbor]))

    return None, 0, n_exp



# So sanh A* va UCS
path_a, cost_a, exp_a = astar(graph, heuristic, 'S', 'G')
path_u, cost_u, exp_u = ucs(graph, 'S', 'G')

print("=" * 55)
print("SO SANH A* vs UCS")
print("=" * 55)
print(f"{'Thuat toan':<12} {'Duong di':<22} {'Cost':<6} {'#Expand':<10}")
print("-" * 55)
print(f"{'A*':<12} {' -> '.join(path_a):<22} {cost_a:<6} {exp_a:<10}")
print(f"{'UCS':<12} {' -> '.join(path_u):<22} {cost_u:<6} {exp_u:<10}")

# TODO (comment): Nhan xet
# - Hai thuat toan co cung tong chi phi khong?
#Có, cả hai đều tìm được đường đi tối ưu với chi phí là 8
#---------------------------------------------------
# - Thuat toan nao expand it node hon? Vi sao?
#A* expand ít node hơn (4 so với UCS thường là 6 hoặc 7, vì A* có heuristic dẫn đường)
#giúp tập trung vào các node có hướng về đích thay vì expand đều mọi hướng như UCS
#--------------------------------------------------
# - Khi nao A* va UCS cho ket qua giong nhau?
#Khi heuristic h(n) = 0 với mọi node, lúc đó f9n) = g(n), A* trở thành UCS
````

````python
# CAU 3a – Nhin anh phia tren, go lai toan bo code va dien 7 cho trong
# ============================================================
import copy, math

X, O, EMPTY = "X", "O", None

# TODO: Go lai va dien 7 cho trong
def initial_state():
  return [[EMPTY]*3 for i in range(3)]

def player(board):
  x_count = sum(row.count(X) for row in board)
  o_count = sum(row.count(O) for row in board)
  return X if x_count == o_count else O #Trong1

def actions(board):
  return {(i, j) for i in range(3) for j in range(3)
  if board [i][j] == EMPTY} #Trong2

def result (board,action):
  i, j = action
  if board [i][j] is not EMPTY: raise ValueError("Invalid action")
  new_board = copy.deepcopy(board)
  new_board[i][j] = player(board) #Trong3
  return new_board

def winner(board):
  for p in [X,O]:
    for row in board:
      if all(c == p for c in row):return p
    for col in range(3):
      if all(board[r][col]==p for r in range(3)): return p
      if all(board[i][i]==p for i in range(3)): return p #Trong4
  return None

def terminal(board):
  return winner(board) is not None or not actions(board)

def utility(board):
  w = winner(board)
  return 1 if w == X else -1 if w == O else 0

def max_value(state):
  if terminal(state): return utility(state)
  v = -math.inf
  for a in actions(state):
    v = max(v, min_value(result(state, a))) #Trong5
  return v

def min_value(state):
    if terminal(state): return utility(state)
    v = math.inf
    for a in actions(state):
        v = min(v, max_value(result(state, a)))
    return v

def minimax(board):
  if terminal(board): return None
  if player(board)==X:
    return max(actions(board),
               key=lambda a: min_value(result(board, a)))

  else:
    return min(actions(board),
               key=lambda a: max_value(result(board, a))) #Trong7
````

````python
# Helper de in bang
def print_board(board):
    print("  0 1 2")
    for i, row in enumerate(board):
        print(f"{i} " + " ".join(c if c else '.' for c in row))
    print()

# Kiem tra cau 3b - state #1
state_1 = [
    [X, X, EMPTY],
    [O, O, EMPTY],
    [EMPTY, EMPTY, EMPTY]
]
print("STATE #1 (luot X):")
print_board(state_1)
print(f"Minimax chon: {minimax(state_1)}   (mong doi: (0,2) - X thang ngay hang 0)")
print()

# Kiem tra cau 3c - state #2
state_2 = [
    [X, X, EMPTY],
    [EMPTY, O, EMPTY],
    [EMPTY, EMPTY, EMPTY]
]
print("STATE #2 (luot O):")
print_board(state_2)
print(f"Minimax chon: {minimax(state_2)}   (mong doi: (0,2) - chan X thang hang 0)")
````


Ô trống: `(0,2), (1,2), (2,0), (2,1), (2,2)`

Phân tích từng nước đi của X:
- **X đi `(0,2)`** → hàng 0 = `X,X,X` → **X thắng ngay** → utility = ?
- **X đi `(1,2)`** → chặn O (vì O đe dọa hàng 1). Game tiếp tục, kết quả = ?
- **X đi `(2,0)`, `(2,1)`, hoặc `(2,2)`** → KHÔNG chặn O → O sẽ đi `(1,2)` → hàng 1 = `O,O,O` → **O thắng** → utility = ?

X là MAX → chọn nước có `min_value` LỚN nhất.  
**Kết luận:** Minimax chọn ô **(0,2)**

**Câu 3c – Tính tay state #2 (lượt O):**

Bảng:
```
X | X | _
_ | O | _
_ | _ | _
```

X đang đe dọa thắng tại ô **___** (X có 2 quân liên tiếp ở hàng/cột/chéo nào?).

Nếu O **không chặn**, lượt sau X sẽ thắng → utility = +1 → tệ cho O.  
Vậy O (MIN) phải chọn ô để **chặn** X.

**Kết luận:** Minimax (O) chọn ô **(0,2)**

````python
# CAU 3d – Cai dat vong lap game (nguoi vs AI)
# ============================================================

def play_game():
    """
    Vong lap game: Nguoi (X) vs AI (O).
    - Nguoi nhap toa do dang 'hang cot' (vd: '1 2')
    - AI dung minimax() chon nuoc di toi uu
    - In bang sau moi luot, kiem tra ket thuc
    """
    board = initial_state()
    print("=" * 40)
    print("TICTACTOE – Nguoi (X) vs AI (O)")
    print("Nhap toa do dang: hang cot (vd: 0 1 hoac 1 2)")
    print("=" * 40)
    print_board(board)

    while not terminal(board):
        current = player(board)

        if current == X:
            # TODO: Nhan input nguoi choi, parse 'hang cot', validate
            valid_move = False
            while not valid_move:
                try:
                    user_input = input("Luot cua X (nhap hang cot): ")
                    # Parse chuoi nhap vao thanh 2 so nguyen
                    i, j = map(int, user_input.split())

                    # Validate: kiem tra trong gioi han va o do phai con trong (EMPTY)
                    # Ghi chu: Ban co the thay the dieu kien nay bang: if (i, j) in actions(board):
                    if 0 <= i <= 2 and 0 <= j <= 2 and board[i][j] != X and board[i][j] != O:
                        action = (i, j)
                        board = result(board, action)
                        valid_move = True
                    else:
                        print("Toa do khong hop le hoac o da bi chiem! Vui long nhap lai.")
                except ValueError:
                    print("Dinh dang sai! Vui long nhap 2 so cach nhau boi dau cach (vd: 0 2).")

        else:
            # TODO: AI dung minimax()
            print("AI (O) dang suy nghi...")
            action = minimax(board)
            print(f"AI di o: {action[0]} {action[1]}")
            board = result(board, action)

        print_board(board)

    # TODO: In ket qua: 'X thang!' / 'O thang!' / 'Hoa!'
    # Ban co the dung ham winner(board) neu da khai bao, hoac dung utility(board)
    win_player = winner(board)
    if win_player == X:
        print("Ket thuc: X thang!")
    elif win_player == O:
        print("Ket thuc: O thang!")
    else:
        print("Ket thuc: Hoa!")

# Bo comment dong duoi de chay (tat khi nop bai)
# play_game()
````

1. Lý do Minimax chậm:

2. Bốn cách cải tiến chính:

Cắt tỉa Alpha-Beta: Dừng duyệt và cắt bỏ các nhánh con chắc chắn đem lại kết quả xấu, giúp thuật toán chạy nhanh hơn hẳn mà vẫn đảm bảo chọn đúng nước đi tốt nhất.

Giới hạn độ sâu & Hàm đánh giá (Heuristic): Ép máy tính chỉ nhìn trước một vài bước (ví dụ: 4-5 bước). Tại bước đó, dùng một hàm chấm điểm tự chế để ước lượng lợi thế hiện tại, thay vì bắt máy tính tính đến tận lúc hết game.

Lưu trữ trạng thái (Memoization / Bảng băm): Ghi nhớ lại điểm số của các thế cờ đã tính toán qua. Nếu bàn cờ lặp lại thế cờ đó, chỉ việc lấy kết quả ra dùng luôn chứ không tính lại từ đầu.
