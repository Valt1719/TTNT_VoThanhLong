```python
#11 tạo mảng a với n phần tử
n=int(input("Nhập số lượng phần tử:"))
a=[]
for i in range(n):
  temp=int(input(f"Nhập phần tử thứ {i+1}:"))
  a.append(temp)
b=[]
for x in a:
  if x % 2 ==0:
    b.append(x)
print("Phần tử của a:", a)
print("Mảng b chứa phần tử chẵn của a:",b)
```

```python
#12 Ch trình tính tổng các số 0- 999 là bội 3 hoặc 5

tong = 0
for i in range(1000):
    if i % 3 == 0 or i % 5 ==0:
        tong += i
        print(f"Tong cac so la boi cua 3 va 5 tu 0 den 99 la: {tong}")

```
