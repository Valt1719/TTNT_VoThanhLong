```python
#13 Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của
# mảng mới là tổng của 2 phần tương ứng từ 2 mảng cho trước.
# Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại
# của mảng đó sẽ đưa vào mảng mới
def Tron_hai_mang(a, b):
  ketqua= []
  min_len = min(len(a), len(b))

  for i in range(min_len):
    ketqua.append(a[i]+b[i])

  if len(a) > len(b):
    ketqua.extend(a[min_len:])
  else:
    ketqua.extend(b[min_len:])
  return ketqua
a=[3, 9, 1, 4]
b=[2, 7, 4, 3, 2, 8]
print(Tron_hai_mang(a,b))

```

```python
#14 Mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
#Tạo mạng a chứa các số a ngẫu nhiên
import random
m=int(input('Nhập số dòng cho mảng:'))
n=int(input('Nhập số cột cho mảng:'))
a=[]

import numpy as np
mang = np.random.randn(m,n).astype(int)
mang

# b Xuất các phần tử thuộc dòng k
k=int(input('Nhập số dòng k:'))
if 0 <= k < mang.shape[0]:
  dong_k = mang[k]
  print(f'Các phần tử dòng k:{k}: {dong_k}')
else:
  print('k thấy dòng k')

# c.Xuất các phần tử cột l
l=int(input('Nhập số cột l:'))
if 0 <= l < mang.shape[1]:
  cot_l=mang[:, l]
  print(f'Các phần tử cột l:{l}: {cot_l}')
else:
  print('k thấy dòng l')

# d.Tìm dòng có tổng lớn nhất ()
import numpy as np
tong_cac_dong = np.sum(mang, axis=1)  # Tính tổng của tất cả các dòng cùng lúc
vitri_max = np.argmax(tong_cac_dong) # Tìm chỉ số (index) của dòng có tổng lớn nhất
gia_tri_max = tong_cac_dong[vitri_max] # Lấy giá trị tổng lớn nhất đó
print(f"Dòng có tổng lớn nhất là dòng số: {vitri_max}")
print(f"Các phần tử của dòng đó: {mang[vitri_max]}")
print(f"Tổng của dòng đó là: {gia_tri_max}")

#e. Tìm cột có tích bé nhất()
```

```python
#14 Mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
#Tạo mạng a chứa các số a ngẫu nhiên
import random
m=int(input('Nhập số dòng cho mảng:'))
n=int(input('Nhập số cột cho mảng:'))
a=[]
mang_np = np.random.randint(1, 9999999, size=(m, n)) #tối đa tới 9999999
a = mang_np.tolist()
# g. Tr bình cộng các phần tử chẵn thuộc dòng lẻ của a
tong=0
dem=0
for i in range( 1, len(a), 2):
  for so in a[i]:
    if(so % 2 == 0):
      tong+=so
      dem+=1
if dem>0:
  tbc=tong/dem
  print(f"Tbc các phần tử chẵn là: {tbc}")
else:
  print('Ko tìm thấy dòng nào lẻ')
```

```python
#14 Mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
#Tạo mạng a chứa các số a ngẫu nhiên
import random
m=int(input('Nhập số dòng cho mảng:'))
n=int(input('Nhập số cột cho mảng:'))
a=[]
mang_np = np.random.randint(1, 999, size=(m, n)) #tối đa tới 999
a = mang_np.tolist()
#h trung bình cộng các phần tử thuộc biên
TongBien =0
demBien =0
for i in range(m):
  for j in range(n):
      if i == 0 or i == m-1 or j == 0 or j == n-1:
        TongBien+=  a[i][j]
        demBien+=1
if demBien >0:
  tbc = TongBien / demBien
  print(f'Tbc các phần tử biên là:{tbc:.2f}')
```
